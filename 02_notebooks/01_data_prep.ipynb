## Initial Data Observations

- Dataset contains 30,000 customer records with credit card usage behavior and demographic attributes such as education, gender, age, and marital status.
- Six months of payment and bill history provide sufficient historical context for trend analysis.

## Columns Key
- Default payment (1 = Yes, 0 = No).
- Gender (1 = male, 2 = female).
- Education level (1 = graduate school; 2 = university; 3 = high school; 4 = others).
- Marital status (1 = married; 2 = single; 3 = others).
- Monthly payments (-1 = paid on time, 0 = payment made but not fully on time, 1 = one month delay, 2 = two months delay... 9 = nine months delay...etc.).

```python
# Create working copy of the data

df_fe = df.copy()
```

```python
# Create bill amount features

bill_cols = [col for col in df_fe.columns if col.startswith("bill_amt")]
pay_cols = [col for col in df_fe.columns if col.startswith("pay_amt")]

df_fe["avg_bill_amt"] = df_fe[bill_cols].mean(axis=1)
df_fe["max_bill_amt"] = df_fe[bill_cols].max(axis=1)

df_fe["avg_payment_amt"] = df_fe[pay_cols].mean(axis=1)
```

```python
# Payment behavior

df_fe["payment_ratio"] = df_fe["avg_payment_amt"] / df_fe["avg_bill_amt"]
df_fe["payment_ratio"] = df_fe["payment_ratio"].replace([np.inf, -np.inf], np.nan)
```

```python
# Utilization

df_fe["utilization_proxy"] = df_fe["avg_bill_amt"] / df_fe["credit_limit"]
```

```python
# Engagement score

df_fe["engagement_score"] = (
    df_fe["payment_ratio"].fillna(0) * 0.4 +
    (1 - df_fe["utilization_proxy"].clip(0, 1)) * 0.4 +
    (df_fe["avg_bill_amt"] / df_fe["avg_bill_amt"].max()) * 0.2
)
```

```python
# Risk profiling

df_fe["risk_flag"] = df_fe["default"].map({0: "Low Risk", 1: "High Risk"})
```

```python
# Validation check

df_fe[
    ["avg_bill_amt", "payment_ratio", "utilization_proxy", "engagement_score"]
].describe()
```

```python
# Clean bill amounts 
# Bill amounts should not be negative for analysis

df_fe["avg_bill_amt"] = df_fe["avg_bill_amt"].clip(lower=0)
```

```python
# Clean utilization proxy

df_fe["utilization_proxy"] = df_fe["utilization_proxy"].clip(lower=0, upper=1.5)
```

```python
# Clean payment ratio

df_fe["payment_ratio"] = df_fe["payment_ratio"].clip(lower=0, upper=3)
```

```python
# Recalculate engagement score

df_fe["engagement_score"] = (
    df_fe["payment_ratio"].fillna(0) * 0.4 +
    (1 - df_fe["utilization_proxy"]) * 0.4 +
    (df_fe["avg_bill_amt"] / df_fe["avg_bill_amt"].max()) * 0.2
)
```

```python
# Validation check

df_fe[
    ["avg_bill_amt", "payment_ratio", "utilization_proxy", "engagement_score"]
].describe()
```

## Feature Summary

Behavioral and engagement features were developed from raw billing and payment history, key features include:
- Average billing and payment amounts.
- Payment behavior (payment ratio).
- Credit utilization proxy.
- Engagement score.
- Credit risk flag. 

## Fixes
- Negative bill amounts were clipped to zero in order to minimize distortions in utilization & engagement metrics.
- Utilization proxy were clipped to remove extreme noise. With a lower limit of 0 (utilization cannot be negative) and upper limit of 1.5 (allow for realistic over-limit behavior).
- Payment ratio were clipped to minimize noise, as a negative value would be meaningless.
- Recalculate engagement score based on cleaned fields.
